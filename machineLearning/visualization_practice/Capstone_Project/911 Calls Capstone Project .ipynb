```python
import numpy as np
import pandas as pd 
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('911.csv')
```

```python
df['title'].nunique()
```

```python
df['Reasons'] = df['title'].apply(lambda reason: reason.split(':')[0])
```

```python
df['Reasons'].value_counts()
```

```python
sns.countplot(x = 'Reasons', data = df)
```

```python
type(df['timeStamp'].iloc[0])
```

```python
#df['timeStamp'].apply(lambda time: pd.to_datetime(time))
# this will show them as time objects but doesn't set them as time objects
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of Week'].map(dmap)
```

```python
sns.countplot(x = 'Day of Week', data = df, hue = 'Reasons')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
```

```python
sns.countplot(x = 'Month', data = df, hue = 'Reasons')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
```

```python
byMonth = df.groupby('Month').count()
```

```python

byMonth.head()
```

```python
byMonth['twp'].plot()
```

```python
sns.lmplot( x = 'Month', y = 'twp', data = byMonth.reset_index())
```

```python
df['Date'] = df['timeStamp'].apply(lambda time: time.date())
```

```python
byDate = df.groupby('Date').count()['twp'].plot()
plt.tight_layout()
```

```python
df[df['Reasons'] == 'Traffic'].groupby('Date').count()['twp'].plot()
plt.title("Traffic")
plt.tight_layout()
```

```python
df[df['Reasons'] == 'Fire'].groupby('Date').count()['twp'].plot()
plt.title("Fire")
plt.tight_layout()
```

```python
df[df['Reasons'] == 'EMS'].groupby('Date').count()['twp'].plot()
plt.title("EMS")
plt.tight_layout()
```

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reasons'].unstack()
dayHour.head()
```

```python
plt.figure(figsize = (12, 6))
heat = sns.heatmap(dayHour, cmap= 'viridis', linecolor = 'white', lw = 1)
heat.set_yticklabels(heat.get_yticklabels(), rotation = 0)
heat.set_xticklabels(heat.get_xticklabels(), rotation = 0)
```

```python
cluster = sns.clustermap(dayHour, cmap = 'viridis')
```

```python
cluster = sns.clustermap(dayHour, cmap = 'viridis', metric = 'correlation')
```

```python
dayMonth = df.groupby(by = ['Day of Week', 'Month']).count()['Reasons'].unstack()
dayMonth.head()
```

```python
plt.figure(figsize = (12, 6))
month_heatmap = sns.heatmap(dayMonth, cmap = 'viridis')
month_heatmap.set_yticklabels(month_heatmap.get_yticklabels(), rotation = 0)
month_heatmap.set_xticklabels(month_heatmap.get_xticklabels(), rotation = 0)
```

```python
month_cluster = sns.clustermap(dayMonth, cmap = 'viridis')
```
