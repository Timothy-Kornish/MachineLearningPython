```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)
```

```python
# Bank of America
BAC = data.DataReader("BAC", 'google', start, end)
#BAC is name of Company
#google is for google finance

# CitiGroup
C = data.DataReader("C", 'google', start, end)

# Goldman Sachs
GS = data.DataReader("GS", 'google', start, end)

# JPMorgan Chase
JPM = data.DataReader("JPM", 'google', start, end)

# Morgan Stanley
MS = data.DataReader("MS", 'google', start, end)

# Wells Fargo
WFC = data.DataReader("WFC", 'google', start, end)
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
# columns axis = 1
#row axis = 0
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis = 1, keys = tickers)
```

```python
bank_stocks
```

```python
bank_stocks.columns.names = ['Bank Ticker', 'Stock Info']
```

```python
bank_stocks.head()
```

```python
#Max closing stock value for every bank
for tick in tickers:
    print(tick, bank_stocks[tick]['Close'].max())
```

```python
#Max closing stock value for every bank using pandas .xs "cross-section"
bank_stocks.xs(key = 'Close', axis = 1, level = 'Stock Info').max()
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick + ' Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
returns.head()
```

```python
# high standard deviation of stock means it's pretty risky, low std dev means pretty steady stock
returns.std()
```
