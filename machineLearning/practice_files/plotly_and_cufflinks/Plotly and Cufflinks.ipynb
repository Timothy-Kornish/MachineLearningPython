```python
df1 = pd.DataFrame(np.random.randn(100, 4), columns = 'A B C D'.split())
df2 = pd.DataFrame({"Category": ['A', 'B', 'C'], "Values": [32, 43, 50]})
```

```python
#Shows line plot of dataframe
df1.plot()
```

```python
# iplot makes plot interactive
#can zoom in, double click to zoom back out
df1.iplot()
```

```python
df1.iplot(kind = 'scatter', x = 'A', y = 'B', mode = 'markers', size = 15)
```

```python
df2.iplot(kind = 'bar', x = 'Category', y = 'Values' )
```

```python
df1.count().iplot(kind = 'bar')
#all culoumns have 100 rows, so counts should be 100
```

```python
df1.sum().iplot(kind = 'bar')
```

```python
df1.iplot(kind = 'box')
#can hide boxes by clicking on that category in the legend
```

```python
# 3-D surface plot
df3 = pd.DataFrame({'x': [1,2,3,4,5], 'y':[10,20,30,20,10], 'z':[500,400,300,200,100]})
```

```python
df3
```

```python
df3.iplot(kind = 'surface', colorscale = 'rdylbu')
```

```python
# histogram of one column in data frame
df1['A'].iplot(kind = 'hist', bins = 30)
```

```python
# histogram of all columns in data frame, ovelapping, can hide columns by clicking the legend
df1.iplot(kind = 'hist')
```

```python
#good for stock data, to show spread between stock values
df1[['A', 'B']].iplot(kind = 'spread')
```

```python
# bubble plots often used for population factors, seen in UN reports
df1.iplot(kind = 'bubble', x = 'A', y = 'B', size = 'C')
```

```python
#will load slow with large data sets, might crash the kernal if too large
df1.scatter_matrix()
```
